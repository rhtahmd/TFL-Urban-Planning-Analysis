```python
# import modules
import pandas as pd
```

```python
# read in csv file as a pandas dataframe
df = pd.read_csv('C:\\TFL Urban Planning Project\\Nov09JnyExport.csv')
```

```python
# explore head of data
print(df.head())
```

```python
# count the number of cells in 'EntTimeHHMM' column that contains '00:00'
entcount = df[df['EntTimeHHMM'] == '00:00'].shape[0]
print(entcount)
```

```python
# count the number of cells in 'EXTimeHHMM' column that contains '00:00'
extcount = df[df['EXTimeHHMM'] == '00:00'].shape[0]
print(extcount)
```

```python
# remove rows that contain EntTimeHHMM and EXTimeHHMM which have the values '00:00'

df = df[(df['EntTimeHHMM'] != '00:00') & (df['EXTimeHHMM'] != '00:00')]
```

```python
# remove busses from the data set as the data is incomplete
df = df.drop('RouteID', axis=1)
df = df[~df['SubSystem'].str.contains('LTB')]
```

```python
# i noticed from the data some of the stations say unstarted, i would like to remove those since i want to get a view of the whole journeys, incomplete journeys are of no use
print(sorted(df['StartStn'].unique()))
```

```python
# do the same for end station
print(sorted(df['EndStation'].unique()))
```

```python
# we want to remove 'Unstarted' and 'Unfinished' from our columns StartStn, EndStation
df = df[(df['StartStn'] != 'Unstarted') & (df['EndStation'] != 'Unfinished')]
df.head()
```

```python
# create a journey duration header
df['JourneyDuration'] = df['ExTime'] - df['EntTime']
df.head()
```

```python
# assess the journey duration data for any errors
df = df.sort_values('JourneyDuration', ascending=True)
df
```

```python

# remove all the data with a negative journey duration
df = df[df['JourneyDuration'] > 0]

df = df.sort_values('JourneyDuration', ascending=True)
df
```

```python
#remove all data with those who tap in and out of the same station
df = df[df['StartStn'] != df['EndStation']]
```

```python
# Assess the data based on journey duration, I want to see as very short journeys could be errors
df = df.sort_values('JourneyDuration', ascending=True)

df
```

```python
# checking for journey durations which equate to 1 and above to check for errors
df[df['JourneyDuration'] >= 1]
df
```

```python
# now i want to remove where people tapped in and out of the station Balham SCL and Balham NR
df = df[(df['StartStn'] != 'Balham NR') | (df['EndStation'] != 'Balham SCL')]
df = df[(df['StartStn'] != 'Balham SCL') | (df['EndStation'] != 'Balham NR')]

df[df['JourneyDuration'] >= 1]
df
```

```python
# now i want to remove where people tapped in and out of the station Canary Wharf E2 and Canary Wharf

df = df[(df['StartStn'] != 'Canary Wharf E2') | (df['EndStation'] != 'Canary Wharf')]
df = df[(df['StartStn'] != 'Canary Wharf') | (df['EndStation'] != 'Canary Wharf E2')]

df[df['JourneyDuration'] >= 1]
df
```

```python
# now i want to remove where people tapped in and out of the station Liverpool St NR and Liverpool St WAGN TOC Gates

df = df[(df['StartStn'] != 'Liverpool St NR') | (df['EndStation'] != 'Liverpool St WAGN TOC Gates')]
df = df[(df['StartStn'] != 'Liverpool St WAGN TOC Gates') | (df['EndStation'] != 'Liverpool St NR')]

df[df['JourneyDuration'] >= 1]
df
```

```python
# now i want to remove where people tapped in and out of the station Liverpool St NR and Liverpool St WAGN TOC Gates

df = df[(df['StartStn'] != 'Fenchurch St NR') | (df['EndStation'] != 'FENCHURCH ST NR')]
df = df[(df['StartStn'] != 'FENCHURCH ST NR') | (df['EndStation'] != 'Fenchurch St NR')]

df[df['JourneyDuration'] >= 1]
df
```

```python
df = df[~df['SubSystem'].str.contains('HEX')]
df
```

```python
# Saving the modified dataset to a new CSV file
df.to_csv('modified', index=False)
```
